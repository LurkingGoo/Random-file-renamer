```python
import os
import random
import string
import shutil
current_directory = os.getcwd()
print(f"Current directory: {current_directory}")
def generate_random_name():
    random_chars = ''.join(random.choices(string.ascii_letters + string.digits, k=10))
    return f"file_{random_chars}"

def bulk_file_rename(directory_path):
    if not os.path.isdir(directory_path):
        print(f"Error: '{directory_path}' is not a valid directory.")
        return
    
    os.chdir(directory_path)
    file_list = os.listdir()
    existing_names = set()

    for file_name in file_list:
        if os.path.isfile(file_name):
            file_extension = os.path.splitext(file_name)[1]
            new_file_name = generate_random_name() + file_extension

            for _ in range(10):
                if new_file_name not in existing_names:
                    existing_names.add(new_file_name)
                    break
                else:
                    new_file_name = generate_random_name() + file_extension
            
            os.rename(file_name, new_file_name)
            print(f"Renamed '{file_name}' to '{new_file_name}'")

# Replace 'directory_path' with the path where your files are located
data_dir = r'C:\Users\lucas\Downloads\red'

bulk_file_rename(data_dir)

```
